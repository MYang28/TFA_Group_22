```python
import numpy as np
import pandas as pd
```

```python
#Reading the csv file and setting it to a variable
data = pd.read_csv('311_Service_Requests_2020.csv')
data
```

```python
#Selected 10025 as the our zipcode, change the type from float to int, and set it to a variable.
df1 = data[data['Incident Zip']== 10025.0]
df1['Incident Zip'] = df1['Incident Zip'].astype(int)
df1.info()
```

```python
#Looking at the unique attributes in the Complaint Type and identifying "Illegal Parking"
df1['Complaint Type'].unique()
```

```python
#First the total fraction of parking incidents were calculated for ALL Zip codes
```

```python
#Counting all incidents under 'Complaint Type'. This will serve as the denominator in the fraction.
total_incident = data['Complaint Type'].count()
total_incident
```

```python
#Counting each unique 'Complaint Type' and will only take the value for Illegal Parking which will serve as the numerator in the fraction.
total_parking = data['Complaint Type'].value_counts()
total_parking
```

```python
#calculating the percentage of illegal parking for all zipcodes
total_fraction = (total_parking['Illegal Parking']/total_incident)*100
total_fraction
#The percentage of illegal parking for all zip codes is 7.4873
```

```python
#Calculate the fraction for illegal parking only in the 10025 Zip code
```

```python
#Counting all incidents under 'Complaint Type'. This will serve as the denominator in the fraction.
total_incident_zip = df1['Complaint Type'].count()
total_incident_zip
```

```python
#Counting each unique 'Complaint Type' and will only take the value for Illegal Parking which will serve as the numerator in the fraction.
total_parking_zip = df1['Complaint Type'].value_counts()
total_parking_zip
```

```python
#calculating the percentage of illegal parking for 10025 zip codes
total_fraction_zip = (total_parking_zip['Illegal Parking']/total_incident_zip)*100
total_fraction_zip 
#The percent of Illegal parking for 10025 zip codes is 3.7754
```

```python
#Creating a Boolean to return false if the fraction for 10025 zip is less than all zip codes and true if else
higher_parking_proportion = bool()
if(total_fraction_zip < total_fraction):
    higher_parking_proportion = False
else:
    higher_parking_proportion = True
higher_parking_proportion
```
